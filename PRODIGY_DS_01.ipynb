# **Create a bar chart or a histogram to visualize the distribution of categorical or continuous variables, such as the distribution of ages or genders in a population.**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('/content/titanic_dataset.csv')
```

```python
data.head()
```

we will be using the titanic dataset to plot the graphs of the number of males and females that survived.

```python
#First we plot the barplot.
sns.barplot(x="Sex",y="Survived",hue="Sex",data=data)
plt.show()
```

```python
#Now we plot the histogram
sns.histplot(data=data,x="Age",y="Survived",hue="Sex")
plt.show()
```

```python
sns.histplot(data=data, x="Age",y="Survived", hue="Sex", multiple="stack")
plt.show()
```
